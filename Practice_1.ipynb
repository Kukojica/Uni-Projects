```python
import pandas as pd #загрузка библиотеки для работы с данными
```

```python
data_t = pd.read_excel('Python. АДвМ Задания.xlsx') #создание датасета с данными о прохождении практических заданий
data_test = pd.read_excel('Python. АДвМ КР.xlsx') #создание датасета с данными о прохождении контрольных работ
```

## Подсчет студентов, кто прошёл на 100 баллов 65% заданий

```python
data_t.head() #отображение первых 5 строк датасета с заданиями
```

```python
data_t.columns = data_t.columns.str.replace('Виртуальная лаборатория программирования:', '')
#коррекция названий столбцов для более удобной работы с ними
data_t.columns = data_t.columns.str.replace(' (Значение)', '', regex=False)
```

```python
data_t.info() #общая информация о данных в датасете
```

```python
data_tf = data_t.copy() #создание копии датасета для дальнейшей работы
for i in data_tf.columns: #цикл, который проходит по всем столбцам датасета
    data_tf.loc[data_tf[i] == '-', [i]] = None #замена всех знаков дефис, поскольку они не поддаются обработке
data_tf = data_tf.dropna(axis=1, how='all') #удаление пустых столбцов, которые были случайно загружены вместе с другими
data_tf_2 = data_tf.copy() #копирование датасета для дополительного анализа
```

```python
data_tf.head() #первые 5 строк форматированного датасета
```

```python
cols = data_tf.columns[3:] #ограничим столбцы только теми, которые содержат оценки по заданиям
```

```python
round(len(cols) * 0.65) #подсчёт числа заданий, которок необходимо завершить для преодоления порога в 65%
```

```python
data_tf['completed_t'] = data_tf[cols].count(axis=1) 
#посчитаем для каждого студента число завершенных заданий
#вне зависимости от оценки
```

```python
data_tf['success_t'] = data_tf[data_tf[cols] == 100].count(axis=1) #посчитаем число успешно сданных заданий
#т.е. сданных на 100 баллов
```

```python
data_tf['perfect_t_percent'] = round((data_tf['success_t'] / 77) * 100)
#посчитаем процент успешно завершенных заданий от общего числа доступных заданий
```

```python
data_tf = data_tf.query('success_t >= 50') #выберем только тех студентов, которые прошли порог в 65% успешно сданных заданий 
```

```python
data_tf = (data_tf[['Фамилия', 'Имя', 'Адрес электронной почты', 'completed_t', 'success_t', 'perfect_t_percent']]
 .sort_values(by='perfect_t_percent',ascending=False)
 .reset_index(drop=True)) #пересоздание индексов для удобной работы
#отсортируем полученный датасет по убыванию процента успешно завершенных заданий
```

```python
data_tf #полученные данные
```

## Подсчет студентов, кто прошёл на 65% контрольные вопросы по всем темам

```python
data_test.info() #основная информация о датасете с контрольными заданиями
```

```python
data_test.columns = data_test.columns.str.replace('Тест:', '').str.replace(' (Значение)', '', regex=False)
data_test.columns = data_test.columns.str.replace(' ', '_')
#коррекция назаваний столбцов для более удобной работы с ними
```

```python
data_test.head() #первые 5 строк датасета
```

```python
cols_test = data_test.columns #создание списка с названиями столбцов
```

```python
for i in data_test.columns:
    data_test.loc[data_test[i] == '-', [i]] = None
#замена в цикле знаков дефис на None для обработки датасета
```

```python
data_test = data_test.dropna(axis=1, how='all') #удалим пустые, столбцы, которые могли случайно попасть в выгрузку
```

```python
cols_test = data_test.columns #новый список с названиями столбцов
```

```python
data_test['completed_test'] = data_test[cols_test[3:]].count(axis=1) #подсчёт общего числа завершенных тестов
#вне зависимости от оценки
```

```python
data_test[cols_test[3:]] = data_test[cols_test[3:]].fillna(0) #замена всех пропусков на 0,
#поскольку иначе невозможно посчитать оценки
data_test[cols_test[3:]] = data_test[cols_test[3:]].astype('float') #изменение типа данных в датасете на вещественный
```

```python
data_test.head() #первые 5 строк полученного датасета
```

```python
data_test['success_test_5'] = data_test[data_test[cols_test[4:7]] >= 3.25].count(axis=1)
#подсчёт успешно завершенных тестов с макс.оценкой в 5 баллов (65% правильных ответов - балл выше 3.25)
data_test['success_test_10'] = data_test[data_test[cols_test[[3,7,8,9,10]]] >= 6.5].count(axis=1)
#подсчёт успешно завершенных тестов с макс.оценкой в 10 баллов (65% правильных ответов - балл выше 6.5)
data_test['success_total'] = data_test['success_test_10'] + data_test['success_test_5']
#объединение полученных значений для подсчёта общего числа успешных тестов
```

```python
data_test.head() #первые 5 строк полученного датасета
```

```python
data_test_f = (data_test.query('success_total == 8')
               [['Фамилия', 'Имя', 'Адрес_электронной_почты', 'success_total']]
               .reset_index(drop=True)) #выборка из студентов, которые успешно сдали все тесты (оценка не менее 65%)
```

```python
data_test_f #полученный датасет
```

**Комментарий:** В список вошло 30 студентов, которые успешно сдали все тесты (*success_total*)

```python
fav = pd.merge(data_test_f,
               data_tf[['Фамилия', 'Имя', 'completed_t','success_t','perfect_t_percent']],
               how='inner',
               on=['Фамилия', 'Имя'])
#объединение датасета со списком студентов, успешно сдавших все практические задания
#и датасета со списком студентов, успешно сдавших все тесты
```

```python
fav #полученный датасет
```
